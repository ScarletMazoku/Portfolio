#### 7. Найдите курсы по выбору, на которые записывались как студенты 18-го года набора, так и студенты 17-го года (0.5 балла)

````python
# Уникальные значения курсов
courses_18_group_unique = pd.DataFrame(pd.unique(df[df['18_group'] != 0][courses_list].values.ravel()))
courses_17_group_unique = pd.DataFrame(pd.unique(df[df['17_group'] != 0][courses_list].values.ravel()))

# Не учитываем тех, кто никуда не записался
courses_18_group_unique = courses_18_group_unique[courses_18_group_unique[0] != ' ']
courses_17_group_unique = courses_17_group_unique[courses_17_group_unique[0] != ' ']

# Мержим и получаем пересечение курсов студентов 17-ого и 18-ого годов
courses_17_and_18_group = courses_17_group_unique.merge(courses_18_group_unique, on = 0,how='inner')
courses_17_and_18_group
````

Методом исключения найдите курсы, которые предлагались только студентам 18-го года и только студентам 17-го года.

````python
# Опять же мержим, исключая те курсы, которые предлагались 18 году, и получаем курсы которые предлагались только студентам 17-го года
courses_17_group_unique.merge(courses_17_and_18_group, how='left', indicator=True) \
          .query("_merge == 'left_only'") \
          .drop('_merge', axis=1)
````

````python
# Опять же мержим, исключая те курсы, которые предлагались 17 году, и получаем курсы которые предлагались только студентам 18-го года
courses_18_group_unique.merge(courses_17_and_18_group, how='left', indicator=True) \
          .query("_merge == 'left_only'") \
          .drop('_merge', axis=1)
````

### Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. Здесь будут описаны ключевые правила оформления графиков для **всех** домашних заданий.

У matplotlib, конечно же, есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторой величины;
- scatter — график, показывающий взаимосвязь двух величин;

В этом задании вы попробуете построить один из них. Не забывайте про базовые принципы построения приличных графиков:
- оси должны быть подписаны, причём не слишком мелко;
- у графика должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах);
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать;
- если распределение на гистограмме имеет тяжёлые хвосты, лучше использовать логарифмическую шкалу.

### 2. Распределение студентов по курсам.

Теперь вам нужно распределить студентов по осенним курсам по выбору, учитывая их предпочтения.

Алгоритм распределения студентов по курсам:
1. По умолчанию на каждой дисциплине по выбору у 3 и 4 курсов может учиться 1 группа (до 30 студентов). Исключения описаны ниже. На blended-дисциплинах для четверокурсников количество мест не ограничено.
2. Проводится первая волна отбора. Для каждой дисциплины формируется список тех, кто указал её первым приоритетом (если студент должен выбрать два курса по выбору, то для него дисциплины, которые он указал первым и вторым приоритетом, рассматриваются как дисциплины первого приоритета). Если желающих больше, чем мест, то выбирается топ по перцентилю рейтинга.
3. На дисциплинах, где остались места после первой волны, формируются списки тех, кто выбрал их вторым приоритетом, и места заполняются лучшими по перцентили рейтинга студентами. После этого проводится такая же процедура для дисциплин третьего приоритета.
4. Если студент не попал на необходимое количество курсов по итогам трёх волн, с ним связывается учебный офис и решает вопрос в индивидуальном порядке.

Обращаем ваше внимание на следующие детали:

- По умолчанию студент выбирает один осенний и один весенний курс по выбору, а также четверокурсники выбирают один blended-курс. Студенты групп 17-го года специализаций МОП и ТИ выбирают по 2 осенних и 2 весенних курса по выбору, также студенты групп 18' специализации МИ выбирают 2 осенних курса. <i>Для студентов, которые выбирают 2 курса (например, осенних) первый приоритет — <code>fall_1</code> и <code>fall_2</code>, второй приоритет — <code>fall_3</code>. Такие студенты участвуют только в двух волнах отбора</i>.

- Студенты специализации МОП не могут выбрать весенним курсом по выбору Машинное обучение 2. <i>Если студент специализации МОП выбрал Машинное обучение 2, то его приоритеты сдвигаются. Из-за совпадений первого и второго курса по выбору двигать приоритеты не надо</i>.

#### 2. Распределите студентов в соответствии с первым приоритетом (2 балла)

````python
# Определим новые признаки предмета по выбору студентов
df_2_course = df[df['courses_numb'] == 2]
df['course1'] = df['fall_1'].fillna('-')
df['course2'] = df.loc[df_2_course.index, 'fall_2']
df['course2'] = df['course2'].fillna('-')
````

````python
# Выделим список курсов на осень
courses_list_fall = ["fall_1", "fall_2",'fall_3']

# Уникальные значения курсов
courses_group_unique = pd.unique(df[courses_list_fall].values.ravel())

# Создаем словарь количества выбранных курсов
zeros_array = np.zeros(12, dtype=int)
dict_courses = dict(zip(courses_group_unique, zeros_array))
````

````python
# Кладем в словарь количество выбранных курсов
for index, item in df['course1'].items():
  if item != '-':
    val = dict_courses.get(item)
    dict_courses.update({item:val+1})

for index, item in df['course2'].items():
  if item != '-':
    val = dict_courses.get(item)
    dict_courses.update({item:val+1})

dict_courses
````

````python
for key, value in dict_courses.items():

  # Количество мест на каждом курсе
  if key in ['Statistical Learning Theory','Высокопроизводительные вычисления']:
    place_numbers = 60
  elif key == 'Анализ неструктурированных данных':
    place_numbers = float('inf')
  else:
    place_numbers = 30

  # Когда желающих больше, чем мест в группе, сортируем по перцентилю рейтинга и оставляем лучших
  if value > place_numbers:
    df_course1 = df[df['course1'] == key][['percentile']]
    df_course2 = df[df['course2'] == key][['percentile']]
    df_concat = pd.concat([df_course1, df_course2]).sort_values(by='percentile')
    extra_places_index = df_concat[place_numbers:].index


    for index in extra_places_index:
      if index in df_course1.index:
        df.loc[index, 'course1'] = '-'
      if index in df_course2.index:
        df.loc[index, 'course2'] = '-'
````

### Вторая волна

````python
dict_courses_new = dict(zip(courses_group_unique, zeros_array))

for index, item in df['course1'].items():
  if item != '-':
    val = dict_courses_new.get(item)
    dict_courses_new.update({item:val+1})

for index, item in df['course2'].items():
  if item != '-':
    val = dict_courses_new.get(item)
    dict_courses_new.update({item:val+1})

dict_courses_new
````

````python
# Количество мест на каждом курсе
dict_left_places = dict_courses_new.copy()

for key, value in dict_courses_new.items():
  # print(value)
  if key in ['Statistical Learning Theory','Высокопроизводительные вычисления']:
    place_left = 60 - value
  elif key == 'Анализ неструктурированных данных':
    place_left = 100000
  else:
    place_left = 30 - value

  if place_left > 0:
    dict_left_places.update({key:place_left})
````

````python
# Новые признаки предмета по выбору студентов
df_1_course = df[(df['courses_numb'] == 1) & (df['course1'] == '-')]
df_2_course = df[(df['courses_numb'] == 2) & ((df['course1'] == '-') | (df['course2'] == '-'))]
````

````python
dict_courses_2 = dict(zip(courses_group_unique, zeros_array))

# Вносим в словарь количество выбранных курсов
for index, item in df_1_course['fall_2'].items():
  if item != '-':
    val = dict_courses_2.get(item)
    dict_courses_2.update({item:val+1})

# Вносим в словарь количество выбранных курсов
for index, item in df_2_course['fall_3'].items():
  if item != '-':
    val = dict_courses_2.get(item)
    dict_courses_2.update({item:val+1})
````

````python
for key, value in dict_courses_2.items():
  if key in dict_left_places.keys():
    left_places_val = dict_left_places.get(key)

    if value > left_places_val:
      df_course1 = df_1_course[df_1_course['fall_2'] == key][['percentile']]
      df_course2 = df_2_course[df_2_course['fall_3'] == key][['percentile']]
      df_concat = pd.concat([df_course1, df_course2]).sort_values(by='percentile')
      extra_places_index = df_concat[:left_places_val].index

      for index in extra_places_index:
        if index in df_course1.index:
          df.loc[index, 'course1'] = key
        if index in df_course2.index:
          if df.loc[index, 'course1'] == '-':
            df.loc[index, 'course1'] = key
          elif df.loc[index, 'course2'] == '-':
            df.loc[index, 'course2'] = key
      else:
        df_course1_index = df_1_course[df_1_course['fall_2'] == key].index
        df_course2_index = df_2_course[df_2_course['fall_3'] == key].index

        for index in df_course1_index:
          df.loc[index, 'course1'] = key

        for index in df_course2_index:
          if df.loc[index, 'course1'] == '-':
            df.loc[index, 'course1'] = key
          elif df.loc[index, 'course2'] == '-':
            df.loc[index, 'course2'] = key
````
