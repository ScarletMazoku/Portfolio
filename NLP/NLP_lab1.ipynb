Скачайте данные отзывов о банковских услугах из папки курса (“data/financial_review.csv”).

```python
filename = "/content/drive/MyDrive/NLP/financial_review.csv"
data = pd.read_csv(filename)
data.head()
```

Выберете один из банков (поле “bank_name”), отфильтуйте отзывы только для рассматриваемого банка. Текстовые данные обрабатываются долго, поэтому можно взять сэмпл данных.

```python
data = data[data['bank_name'] == 'кредит европа банк']
data.head()
```

Выполните всю необходимую предварительную обработку текста (удаление лишних символов, стемминг/лемматизация, токенизация и т.д.).

```python
import pandas as pd
import re

def remove_non_russian(text):
    cleaned_text = re.sub(r'[^а-яА-ЯёЁ]', ' ', text)
    return cleaned_text

data['review'] = data['review'].apply(remove_non_russian)

print(data['review'])
```

```python
data['review'] = data['review'].str.lower()
print(data['review'])
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

stop_words = set(stopwords.words('russian'))

def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

data['review'] = data['review'].apply(remove_stopwords)

print(data['review'])
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize

tokenized_reviews = []

def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens

for index, row in data.iterrows():
    tokens = tokenize_text(row['review'])
    tokenized_reviews.append(tokens)
```

```python
print(tokenized_reviews)
```

```python
tokens = [token for sublist in tokenized_reviews for token in sublist]

tokens = list(set(tokens))
print(tokens)
```

```python
from nltk.stem.snowball import RussianStemmer

stemmer = RussianStemmer()
lemmatized_tokens = [stemmer.stem(token) for token in tokens]
tokens = lemmatized_tokens
print(tokens)
```

Найдите Топ-10 наиболее часто используемых слов в отзывах.

```python
from collections import Counter

all_reviews = ' '.join(data['review'].astype(str))

word_freq = Counter(tokens)

top_words = word_freq.most_common(10)

for word, frequency in top_words:
    print(f"{word}: {frequency}")
```

Разделите отзывы на положительные и отрицательные с использованием значения в поле «score» (например, 1,2- негативные, 4,5-позитивные).

```python
review_pos = data['review'][data['score'].isin(['4', '5'])]
review_neg = data['review'][data['score'].isin(['1', '2'])]

print(review_pos[:5])
print(review_neg[:5])
```

Найдите Топ-10 наиболее часто используемых прилагательных, используемых со словом банк для положительных/отрицательных отзывов.

```python
review_pos = review_pos.tolist()
review_pos = [i for i in review_pos if 'банк' in i]

review_neg = review_neg.tolist()
review_neg = [i for i in review_neg if 'банк' in i]

print(review_pos)
print(review_neg)
```

```python
review_pos = ''.join(review_pos)
print(review_pos)

review_neg = ''.join(review_neg)
print(review_neg)
```

```python
import spacy
nlp = spacy.load("ru_core_news_sm")

doc_pos = nlp(review_pos)
doc_neg = nlp(review_neg)

adjectives_pos = [token.text for token in doc_pos if token.pos_ == "ADJ"]
adjectives_neg = [token.text for token in doc_neg if token.pos_ == "ADJ"]

print(adjectives_pos)
print(adjectives_neg)
```

```python
from collections import Counter

count_pos = Counter(adjectives_pos)
print(count_pos)

count_neg = Counter(adjectives_neg)
print(count_neg)
```

```python
count_pos = count_pos.most_common(10)
count_neg = count_neg.most_common(10)

print('positive')
for word, frequency in count_pos:
    print(f"{word}: {frequency}")
print('-------------------------')
print('negative')
for word, frequency in count_neg:
    print(f"{word}: {frequency}")
```
