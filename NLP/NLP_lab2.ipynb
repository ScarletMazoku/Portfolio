```python
filename = "/content/drive/MyDrive/NLP/financial_review.csv"
data = pd.read_csv(filename)
data.head()
```

```python
data = data[data['bank_name'] == 'кредит европа банк']
data.head()
```

```python
import pandas as pd
import re

def remove_non_russian(text):
    cleaned_text = re.sub(r'[^а-яА-ЯёЁ]', ' ', text)
    return cleaned_text

data['review'] = data['review'].apply(remove_non_russian)

print(data['review'])
```

```python
data['review'] = data['review'].str.lower()
print(data['review'])
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

stop_words = set(stopwords.words('russian'))

def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

data['review'] = data['review'].apply(remove_stopwords)

print(data['review'])
```

2. Постройте векторные представления отзывов с использованием подходов BoW и TF-IDF c разными параметрами (использование n-gramm, различные пороги по отсечению частотных слов), а также с использованием предобучененных векторов word2vec.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

vectorizer = CountVectorizer(ngram_range=(1, 2), stop_words=stopwords.words('russian'), max_df=0.85, min_df=0.1)

X_bow = vectorizer.fit_transform(data['review'])

features_bow = X_bow.toarray()

print("Vocabulary:", vectorizer.get_feature_names_out())

vectorizer_tfidf = TfidfVectorizer(ngram_range=(1, 2), stop_words=stopwords.words('russian'), max_df=0.85, min_df=0.1)

X_tfidf = vectorizer_tfidf.fit_transform(data['review'])

features_tfidf = X_tfidf.toarray()

print("Vocabulary:", vectorizer_tfidf.get_feature_names_out())
```

```python
import gensim.downloader
from gensim.utils import simple_preprocess
import numpy as np

russian_model = gensim.downloader.load('word2vec-ruscorpora-300')

tokenized_reviews = data['review'].apply(lambda x: simple_preprocess(x))

# Создание массива для хранения векторных представлений отзывов
X_word2vec_pretrained = np.zeros((len(tokenized_reviews), russian_model.vector_size))

# Заполнение массива векторными представлениями отзывов
for i, words in enumerate(tokenized_reviews):
    word_vectors = [russian_model[word] for word in words if word in russian_model]
    if word_vectors:
        X_word2vec_pretrained[i] = np.mean(word_vectors, axis=0)

# Вывод размера полученной матрицы векторных представлений
print("Shape of the matrix with pretrained Word2Vec vectors:", X_word2vec_pretrained.shape)
```

3. Разделите отзывы на положительные и отрицательные, можно на основании значения в поле «score» (например, 1,2- негативные, 4,5-позитивные).

```python
review_pos = data['review'][data['score'].isin(['4', '5'])]
review_neg = data['review'][data['score'].isin(['1', '2'])]

print(review_pos[:5])
print(review_neg[:5])
data['sentiment'] = data['score'].apply(lambda x: 'positive' if x in ['4', '5'] else ('negative' if x in ['1', '2'] else 'neutral'))
```

4. Классифицируйте отзывы на положительные/отрицательные с использованием разных векторных представлений. Можно использовать любую модель/модели класссификации.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Создание обучающего и тестового наборов для BoW и TF-IDF
X_train_bow, X_test_bow, y_train_bow, y_test_bow = train_test_split(data['review'], data['sentiment'], test_size=0.2, random_state=42)
X_train_tfidf, X_test_tfidf, y_train_tfidf, y_test_tfidf = train_test_split(data['review'], data['sentiment'], test_size=0.2, random_state=42)

# Обучение BoW модели
vectorizer_bow = CountVectorizer(ngram_range=(1, 2), stop_words=stopwords.words('russian'), max_df=0.85, min_df=0.1)
X_train_bow = vectorizer_bow.fit_transform(X_train_bow)
X_test_bow = vectorizer_bow.transform(X_test_bow)

# Обучение TF-IDF модели
vectorizer_tfidf = TfidfVectorizer(ngram_range=(1, 2), stop_words=stopwords.words('russian'), max_df=0.85, min_df=0.1)
X_train_tfidf = vectorizer_tfidf.fit_transform(X_train_tfidf)
X_test_tfidf = vectorizer_tfidf.transform(X_test_tfidf)

# Обучение логистической регрессии на BoW
model_bow = LogisticRegression()
model_bow.fit(X_train_bow, y_train_bow)
predictions_bow = model_bow.predict(X_test_bow)

# Обучение логистической регрессии на TF-IDF
model_tfidf = LogisticRegression()
model_tfidf.fit(X_train_tfidf, y_train_tfidf)
predictions_tfidf = model_tfidf.predict(X_test_tfidf)
```
